```python
from google_play_scraper import app, Sort, reviews
import numpy as np
import pandas as pd
```

example link application : https://play.google.com/store/apps/details?id=com.tomoro.indonesia.android&hl=id

we get the part after the id= and before the &hl

```python
# Information
?reviews
```

```python
result, continuation_token = reviews(
    'com.tomoro.indonesia.android',
    lang = 'id',
    country = 'id',
    sort = Sort.NEWEST,
    count = 2000,
    filter_score_with = None
)
```

```python
result
```

The extracted data is in list of json format, so we can directly load it into our pandas dataframe by simply store it into dataframe.

```python
# Store and check dataframe
df = pd.DataFrame(result)
df.head()
```

```python
# Checking the lenght whether it is the same as our count parameter or not
len(df)
```
